```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
# %matplotlib qt

font = { 'weight' : 'bold',
        'size'   : 16}
plt.rc('text.latex', preamble=r'\usepackage{amsmath}\usepackage{braket}')
plt.rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
plt.rc('text', usetex=True)
plt.rc('font', **font)
plt.rcParams['savefig.dpi'] = 300
```

# AKLT chain with $\Gamma = 2/3$

```python
N = 64
Gamma = 0.6666666666666666
scalar_fname = f"./src/scalar_E_GS_first_excited_per_bond_dimension_N={N}_Gamma={Gamma}.csv"
obs_fname = f"./src/vec_obs_GS_first_excited_per_bond_dimension_N={N}_Gamma={Gamma}.csv"

df_scalar = pd.read_csv(scalar_fname)
df_obs = pd.read_csv(obs_fname)

fig, ax = plt.subplots(1, 1, figsize=(4,4))

ax.set_xlabel(r"$\chi$", fontsize = 18)
ax.set_ylabel("Energy", fontsize = 18)

ax.plot(df_scalar["bond_dimension"], df_scalar["E0"], "-", color = "green", label = r"$E_0$")
ax.plot(df_scalar["bond_dimension"], df_scalar["E1"], "-", color = "red", label = r"$E_1$")
ax.legend()

fig, ax = plt.subplots(1, 1, figsize=(4,4))

ax.set_xlabel(r"$\chi$", fontsize = 18)
ax.set_ylabel("Energy", fontsize = 18)

ax.plot(df_scalar["bond_dimension"], df_scalar["E1"] - df_scalar["E0"], "-", color = "green", label = r"$\Delta E$")
ax.legend()

fig, ax = plt.subplots(1, 1, figsize=(4,4))

ax.set_xlabel(r"$\chi$", fontsize = 18)
ax.set_ylabel("Energy", fontsize = 18)

ax.plot(df_scalar["bond_dimension"], df_scalar["Svn0"], "-", color = "green", label = r"$S(\ket{\psi_0}\bra{\psi_0})$")
ax.plot(df_scalar["bond_dimension"], df_scalar["Svn1"], "-", color = "red", label = r"$S(\ket{\psi_1}\bra{\psi_1})$")
ax.legend()


observable = "Sz0"

plt.figure(figsize=(8, 5))

for chi in sorted(df_obs["bond_dimension"].unique()):

    subset = df_obs[df_obs["bond_dimension"] == chi]

    values = subset[observable].iloc[0]

    if isinstance(values, str):
        import ast
        values = ast.literal_eval(values)

    sites = np.arange(1, len(values) + 1)

    plt.plot(sites, values, marker='o', linewidth=1.5, label=f"$\\chi={chi}$")

plt.xlabel(r"Site $j$", fontsize=16)
plt.ylabel(r"$S_{j}^{z}$", fontsize=16)
plt.legend()
plt.tight_layout()
plt.show()

observable = "Sz1"

plt.figure(figsize=(8, 5))

for chi in sorted(df_obs["bond_dimension"].unique()):

    subset = df_obs[df_obs["bond_dimension"] == chi]

    values = subset[observable].iloc[0]

    if isinstance(values, str):
        import ast
        values = ast.literal_eval(values)

    sites = np.arange(1, len(values) + 1)

    plt.plot(sites, values, marker='o', linewidth=1.5, label=f"$\\chi={chi}$")

plt.xlabel(r"Site $j$", fontsize=16)
plt.ylabel(r"$S_{j}^{z}$", fontsize=16)
plt.legend()
plt.tight_layout()
plt.show()
```
